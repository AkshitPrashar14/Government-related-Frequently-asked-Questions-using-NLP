```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv("../data/processed/government_qa_clean.csv")
df.shape
```

```python
df.columns
```

```python
df = df[df["final_question"].notna()]
df = df[df["final_question"].str.len() > 20]
```

```python
questions = df["final_question"].astype(str).tolist()
len(questions)

```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
question_embeddings = model.encode(
    questions,
    show_progress_bar=True,
    batch_size=32
)
```

```python
import numpy as np

np.save(
    "../data/processed/question_embeddings.npy",
    question_embeddings
)

```

```python
embeddings = np.load("../data/processed/question_embeddings.npy")
embeddings.shape
```

```python
embeddings = np.load("../data/processed/question_embeddings.npy")
```

```python
import faiss

faiss.normalize_L2(embeddings)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatIP(dimension)
index.add(embeddings)
```

```python
index.ntotal
```

```python
faiss.write_index(
    index,
    "../data/processed/faiss_index.index"
)
```

```python
# Example user query
query = "whether Government has approved any package for FACT Udyogmandal?"

# Embed query
query_embedding = model.encode([query])
query_embedding = query_embedding.astype("float32")
faiss.normalize_L2(query_embedding)

# Search
k = 5
scores, indices = index.search(query_embedding, k)

# Show results
for i, idx in enumerate(indices[0]):
    print(f"\nRank {i+1}")
    print("Score:", scores[0][i])
    print("Question:", df.iloc[idx]["final_question"])
    print("Answer:", df.iloc[idx]["answer"])
```
