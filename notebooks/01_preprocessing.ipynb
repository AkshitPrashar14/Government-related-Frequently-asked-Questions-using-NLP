```python
df = pd.read_csv("../data/raw/govtqa_utf8.csv")
df.head()
```

```python
df.columns 
```

```python
df['final_question'] = (
    df['question_title'].fillna('') + ' ' +
    df['question_description'].fillna('')
)        
```

```python
import re

def clean_question(text):
    text = str(text).lower()
    text = re.sub(r'\(a\)|\(b\)|\(c\)|\(d\)', '', text)
    text = re.sub(r'whether', '', text)
    text = re.sub(r'shri|smt\.?|dr\.', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['question_clean'] = df['final_question'].apply(clean_question) 
```

```python
import re

def clean_answer(text):
    text = str(text)
    text = re.sub(r'minister.*answer', '', text, flags=re.I)
    text = re.sub(r'[\?\*\'\-_=]{3,}', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['answer_clean'] = df['answer'].apply(clean_answer)  
```

```python
df.columns
```

```python
def is_noisy(text, threshold=0.3):
    alnum = len(re.findall(r'[a-zA-Z0-9]', text))
    return alnum / max(len(text), 1) < threshold

df = df[~df['answer_clean'].apply(is_noisy)]
df = df[df['answer_clean'].str.len() > 30]     
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)

df.to_csv(
    "../data/processed/government_qa_clean.csv",
    index=False,
    encoding="utf-8"
)
```
